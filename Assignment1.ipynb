```python
#b.
filtered_df = df[['Genre','global_sales']]
genre = []
labels = []
for item in filtered_df['Genre']:
  if item not in labels:
    labels.append(item)
for item in labels:
  filtered = filtered_df[filtered_df['Genre'].str.contains(f"{item}")]
  genre.append(filtered['global_sales'].sum())

plt.bar(labels,genre)
plt.xticks(rotation = 90)

```
